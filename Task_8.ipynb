```python
import numpy as np
import matplotlib.pyplot as plt
from timeit import default_timer as timer
import pandas as pd
import networkx as nx
from scipy.linalg import lu
```

### Maximum flow

```python
# Initialize function for create graph for maximum flow problem
def create_graph(nodes, neib):
    g = nx.DiGraph(nx.connected_watts_strogatz_graph(nodes, neib, 0.1))
    s = np.random.choice(g.nodes())
    copy = g.copy()
    copy.remove_node(s)
    t = np.random.choice(copy.nodes())
    return gen_weights(g), s, t
```

```python
# Function which add random capacity to graph
def gen_weights(graph):
    local = graph.copy()
    weighted = []
    for i in graph.edges():
        weighted.append(i + (np.random.randint(low=1, high=50),))
    local.add_weighted_edges_from(weighted)
    return local
```

```python
# Create and draw graph
S = nx.connected_watts_strogatz_graph(10, 5, 0.1)
pos = nx.spring_layout(S)
S = nx.DiGraph(S)

S = gen_weights(S)

nx.draw_networkx(S, pos=pos, with_labels=True)
```

```python
# Solve maximum flow problem by preflow-push method and show result
res = nx.maximum_flow(S, 1, 5, capacity='weight')

res
```

```python
# Function for generate preflow-push method timestamps
def make_timestamps_preflow_push(num_it):
    timestamps = []
    nodes = 2
    neib = 2
    for i in range(num_it):
        time = 0
        g, s, t = create_graph(nodes, neib)
        for _ in range(5):
            start = timer()
            nx.maximum_flow(g, s, t, capacity='weight')
            stop = timer()
            time += stop - start
        timestamps.append(time / 5)
        nodes += 1
        if neib <= 5:
            neib += 1
        if nodes > 50 and neib < 10:
            neib += 1
    return timestamps
```

```python
# Function for genereta timestamps for given function
def gen_times(number_of_runs, function):
    timestamps = []
    for run in range(number_of_runs):
        time = 0
        vector = list(np.random.rand(run))
        for _ in range(5):
            start = timer()
            function(vector)
            end = timer()
            time += end - start
        timestamps.append(time / 5)
    return timestamps
```

```python
# Implementation of function with O(n^2)
def bubble_sort(vector):
    for i in range(0,len(vector)-1): 
        for j in range(len(vector)-1): 
            if(vector[j]>vector[j+1]): 
                temp = vector[j]
                vector[j] = vector[j+1] 
                vector[j+1] = temp 
    return vector
```

```python
quadr = gen_times(200, bubble_sort)
```

```python
preflow = make_timestamps_preflow_push(200)
```

```python
# Plot results
x = np.linspace(0, 200, 200)
plt.plot(x, preflow, label='Preflow push method')
plt.plot(x, quadr, label='Function with n^2 Big-O')
plt.ylabel('Second')
plt.xlabel('Size')
plt.title('Comparison Preflow-push method by Big-O notation')
plt.legend()
```

### LUP-decomposition

```python
# Create random matrix
A = np.random.rand(5,5)
# Apply LUP-decomposition
p, l, u = lu(A)
# Check
np.allclose(A, p @ l @ u)
```

```python
# Function for generate LUP-decomposition timestamps
def lu_times(num_it):
    timestamps = []
    for run in range(1, num_it+1):
        time = 0
        matrix = np.random.rand(run, run)
        for _ in range(5):
            start = timer()
            lu(matrix)
            end = timer()
            time += end - start
        timestamps.append(time / 5)
    return timestamps
```

```python
# Implementation of function with O(n^3) 
def gen_matrix_product(num_it):
    timestamps = []
    for i in range(1, num_it+1):
        time = 0
        for _ in range(5):
            A = np.random.rand(i, i)
            B = np.random.rand(i, i)
            start = timer()
            C = A @ B
            end = timer()
            time += (end - start)
        timestamps.append(time / 5)
    return timestamps
```

```python
MAX_ITER = 500
```

```python
lup_dec = lu_times(MAX_ITER)
```

```python
matrix_prod = gen_matrix_product(MAX_ITER)
```

```python
# Plot results
x = np.linspace(0, MAX_ITER, MAX_ITER)
plt.plot(x, lup_dec, label='LUP-decomposition')
plt.plot(x, matrix_prod, label='Function with n^3 Big-O')
plt.title('Comparison LUP-decomposition by Big-O notation')
plt.ylabel('Seconds')
plt.xlabel('Size')
plt.legend()
```
